#### Ejemplos de if, if else, if elif

#### ejemplo numero 1 if
En Python, las estructuras condicionales permiten que el programa tome decisiones y ejecute diferentes bloques de código según se cumplan o no ciertas condiciones.

1. Estas condiciones se construyen utilizando una combinación de operadores relacionales (para comparar valores), lógicos (para combinar condiciones) y aritméticos (para realizar cálculos previos a la comparación)
2. Operadores utilizados en las condiciones
Aritméticos: Permiten realizar cálculos matemáticos como suma (+), resta (-), multiplicación (*), división (/), potencia (**), división entera (//) y módulo (%)
3. Relacionales: Se usan para comparar dos valores y devuelven un resultado booleano (True o False). Incluyen: mayor que (>), menor que (<), igual a (==), diferente de (!=), mayor o igual (>=) y menor o igual (<=)
4. Lógicos: Permiten combinar múltiples expresiones booleanas. Son: and (y), or (o) y not (no).

```python
#1. Ejemplo de if simple Se utiliza para ejecutar un bloque de código únicamente si la condición es verdadera.
edad = int(input("Ingrese su edad: "))
if edad >= 18:
    print("Eres mayor de edad.")
```

```python
#2. Ejemplo de if-else con operadores aritméticos. El bloque else permite manejar el caso en que la condición del if resulte ser falsa.
precio = int(input("Ingrese el precio del producto: "))
descuento = int(input("Ingrese el descuento en porcentaje: "))
if precio - descuento < 95:
    print("Es una oferta.")
else:
    print("Precio regular.")
```

```python
#3. Ejemplo de if-elif-else con operadores lógicos. La estructura elif permite verificar múltiples condiciones de forma sucesiva.
numero = int(input("Ingrese un número: "))
if numero > 10 and numero < 20:
    print("El número está entre 11 y 19.")
elif numero == 10 or numero == 20:
    print("El número es un extremo del rango.")
else:
    print("El número está fuera del rango 10-20.")
```

```python
# 4. Ejemplo integrador: Verificación de año bisiesto
# Este ejemplo utiliza operadores aritméticos (módulo), lógicos y relacionales para resolver un problema lógico complejo
anio = int(input("Ingrese un año: "))
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.")
else:
    print(f"{anio} no es un año bisiesto.")
```
